```python
import numpy as np
import matplotlib.pyplot as plt

# Create a synthetic toy dataset
np.random.seed(0)
X = 2 * np.random.rand(700, 1)
y = 4 + 7 * X + np.random.randn(700, 1)

# Visualize the dataset
plt.scatter(X, y)
plt.xlabel("Input Feature X")
plt.ylabel("Target y")
plt.title("Synthetic Linear Dataset")
plt.show()
```

```python
import tensorflow as tf

# Model parameters
W = tf.Variable(tf.random.normal([1]), name="weight")
b = tf.Variable(tf.random.normal([1]), name="bias")

# Linear regression function
def linear_model(x):
    return W * x + b

# Loss function
def loss_function(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))

# Optimizer
optimizer = tf.optimizers.SGD(learning_rate=0.01)

# Modified training process to record loss
tf_loss_history = []

def train_tf(X, y, epochs=200):
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            y_pred = linear_model(X)
            loss = loss_function(y, y_pred)
        gradients = tape.gradient(loss, [W, b])
        optimizer.apply_gradients(zip(gradients, [W, b]))
        tf_loss_history.append(loss.numpy())

# Perform training
train_tf(X, y)

# Calculate and print the final loss (MSE)
final_pred = linear_model(X)
final_loss = loss_function(y, final_pred).numpy()
print("TensorFlow Model Parameters:")
print("Weight:", W.numpy()[0])
print("Bias:", b.numpy()[0])
print("Final Loss (MSE):", final_loss)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define and train the model
model_sklearn = LinearRegression()
model_sklearn.fit(X, y.ravel())

# Predictions for calculating MSE
predictions_sklearn = model_sklearn.predict(X)
mse_sklearn = mean_squared_error(y, predictions_sklearn)

# Display model parameters and MSE
print("Scikit-learn Model Parameters:")
print("Coefficient:", model_sklearn.coef_[0])
print("Intercept:", model_sklearn.intercept_)
print("Final Loss (MSE):", mse_sklearn)
```

```python
from sklearn.linear_model import SGDRegressor

# Define the SGDRegressor model
model_sgd = SGDRegressor(max_iter=200, tol=1e-3, eta0=0.01, random_state=0)

# Placeholder for loss history
sgd_loss_history = []

# Train the model iteratively to record the loss
for epoch in range(200):
    model_sgd.partial_fit(X, y.ravel())
    y_pred_sgd = model_sgd.predict(X)
    sgd_loss = mean_squared_error(y, y_pred_sgd)
    sgd_loss_history.append(sgd_loss)

# Display model parameters
print("SGDRegressor Model Parameters:")
print("Coefficient:", model_sgd.coef_[0])
print("Intercept:", model_sgd.intercept_[0])
```

```python
# Define the Keras model
model_keras = tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=(1,))])
model_keras.compile(optimizer='sgd', loss='mean_squared_error')

# Train the model and record loss
history_keras = model_keras.fit(X, y, epochs=200, verbose=0)

# Extract loss history
keras_loss_history = history_keras.history['loss']

# Calculate and print the final loss (MSE)
final_loss_keras = model_keras.evaluate(X, y, verbose=0)

# Display model parameters
print("Keras Model Parameters:")
print("Weights:", model_keras.layers[0].get_weights()[0][0][0])
print("Bias:", model_keras.layers[0].get_weights()[1][0])
print("Final Loss (MSE):", final_loss_keras)
```

```python
# TensorFlow predictions over the entire dataset
predictions_tf = linear_model(X).numpy()

# Scikit-learn predictions over the entire dataset
predictions_sklearn = model_sklearn.predict(X)

# Keras predictions over the entire dataset
predictions_keras = model_keras.predict(X)

# Plotting for TensorFlow Model
plt.scatter(X, y, label='Data')
plt.plot(X, predictions_tf, label='TF Predictions', color='green')
plt.xlabel("Input Feature X")
plt.ylabel("Target y")
plt.title("Linear Regression with TensorFlow")
plt.legend()
plt.show()

# Plotting for Scikit-learn Model
plt.scatter(X, y, label='Data')
plt.plot(X, predictions_sklearn, label='Sklearn Predictions', color='red')
plt.xlabel("Input Feature X")
plt.ylabel("Target y")
plt.title("Linear Regression with Scikit-learn")
plt.legend()
plt.show()

# Plotting for Keras Model
plt.scatter(X, y, label='Data')
plt.plot(X, predictions_keras, label='Keras Predictions', color='purple')
plt.xlabel("Input Feature X")
plt.ylabel("Target y")
plt.title("Linear Regression with Keras")
plt.legend()
plt.show()

# Comparing Predictions of All Models
plt.scatter(X, y, label='Data')
plt.plot(X, predictions_tf, label='TF Predictions', color='green')
plt.plot(X, predictions_sklearn, label='Sklearn Predictions', linestyle='dashed', color='red')
plt.plot(X, predictions_keras, label='Keras Predictions', linestyle='dotted', color='purple')
plt.xlabel("Input Feature X")
plt.ylabel("Target y")
plt.title("Comparison of Predictions")
plt.legend()
plt.show()
```

```python
# Loss metrics comparison

# Plotting TensorFlow Loss
plt.plot(tf_loss_history, label='TensorFlow Loss', color='green')

# Plotting Keras Loss
plt.plot(keras_loss_history, label='Keras Loss', color='purple')

# Scikit-learn loss (constant line as it's a final value)
plt.axhline(y=mse_sklearn, label='Scikit-learn Final MSE', color='red', linestyle='dashed')

plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("Loss Metrics Comparison")
plt.legend()
plt.show()
```

```python
# Base model with modifications
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
])
model_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.05), loss='mean_squared_error')
history_1 = model_1.fit(X, y, epochs=100, batch_size=16, verbose=0)
loss_1 = model_1.evaluate(X, y, verbose=0)

# Comment: 
# 1. Increased complexity with additional hidden layers and more neurons.
# 2. Changed optimizer to Adam with a higher learning rate for potentially faster convergence.
# 3. Reduced the batch size and epochs for quicker training.

# Print relevant outputs
weights_1, biases_1 = model_1.layers[-1].get_weights()
print("Model 1 Parameters:")
print("Weights:", weights_1)
print("Biases:", biases_1)
print("Final Loss (MSE):", loss_1)
```

```python
# Model with different activation, learning rate decay, and momentum
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='tanh', input_shape=(1,)),  # Changed activation
    tf.keras.layers.Dense(1)
])
model_2.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, decay=1e-6), 
                loss='mean_squared_error')
history_2 = model_2.fit(X, y, epochs=200, batch_size=32, verbose=0)
loss_2 = model_2.evaluate(X, y, verbose=0)

# Comment: 
# 1. Switched to 'tanh' activation function for a different activation characteristic.
# 2. Introduced learning rate decay and momentum in the SGD optimizer.
# 3. Increased epochs and batch size for more gradual learning.
# Print relevant outputs
weights_2, biases_2 = model_2.layers[-1].get_weights()
print("Model 2 Parameters:")
print("Weights:", weights_2)
print("Biases:", biases_2)
print("Final Loss (MSE):", loss_2)
```

```python
# Model with dropout, different initializer, and different loss function
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', kernel_initializer='he_normal', input_shape=(1,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1)
])
model_3.compile(optimizer='adam', loss='mean_absolute_error')  # Changed loss function
history_3 = model_3.fit(X, y, epochs=150, batch_size=20, verbose=0)
loss_3 = model_3.evaluate(X, y, verbose=0)

# Comment: 
# 1. Added dropout for regularization to prevent overfitting.
# 2. Used 'he_normal' initializer for better performance with ReLU activation.
# 3. Switched to mean absolute error for loss to focus on magnitude of errors.
# Print relevant outputs
weights_3, biases_3 = model_3.layers[-1].get_weights()
print("Model 3 Parameters:")
print("Weights:", weights_3)
print("Biases:", biases_3)
print("Final Loss (MSE):", loss_3)
```

```python
# Model with more layers, different optimizer, and increased learning rate
model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(15, activation='relu'),
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(1)
])
model_4.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.1), loss='mean_squared_error')
history_4 = model_4.fit(X, y, epochs=50, batch_size=10, verbose=0)
loss_4 = model_4.evaluate(X, y, verbose=0)

# Comment: 
# 1. Increased the number of layers and varied the number of neurons for a deeper network.
# 2. Used RMSprop optimizer with a higher learning rate for an aggressive approach.
# 3. Reduced epochs but used a smaller batch size for faster, more frequent updates.


# Print relevant outputs
weights_4, biases_4 = model_4.layers[-1].get_weights()
print("Model 4 Parameters:")
print("Weights:", weights_4)
print("Biases:", biases_4)
print("Final Loss (MSE):", loss_4)
```

```python
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping

# Define the model with L1 regularization
model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', kernel_regularizer=regularizers.l1(0.01), input_shape=(1,)),
    tf.keras.layers.Dense(1)
])
# Compile the model
model_5.compile(optimizer='adam', loss='mean_squared_error')
# Set up early stopping
early_stopping = EarlyStopping(monitor='loss', patience=10)
# Train the model
history_5 = model_5.fit(X, y, epochs=200, batch_size=15, verbose=0, callbacks=[early_stopping])
# Evaluate the model
loss_5 = model_5.evaluate(X, y, verbose=0)

# Print relevant outputs
weights_5, biases_5 = model_5.layers[-1].get_weights()
print("Model 5 Parameters:")
print("Weights:", weights_5)
print("Biases:", biases_5)
print("Final Loss (MSE):", loss_5)
```

```python
# Loss History Comparison
histories = [history_1, history_2, history_3, history_4, history_5]

for i, history in enumerate(histories):
    plt.plot(history.history['loss'], label=f'Model {i+1}')

plt.title("Loss History Comparison")
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.legend()
plt.show()
```
